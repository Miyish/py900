```python
url = 'https://quotes.toscrape.com/'
```

```python
import requests
```

```python
html = requests.get(url).content
html
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
soup.find_all(class_ = 'author')
```

```python
author = [item.text for item in soup.find_all(class_ = 'author')]
author
```

```python
quotes = [item.text for item in soup.find_all(class_ = 'text')]
quotes
```

```python
data = {
    'quotes':quotes,
    'author':author
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('data.csv', index = False)#csv is comma seperated values
```

#Lambda function

1.   nameless function
2.   anonymous function



```python
#syntax# lambda args: return_expression
```

```python
#used with map() and filter()
```

```python
num = [5,6,7]
```

```python
map(lambda item : item ** 2, num)
```

```python
list(map(lambda item : item ** 2, num))
```

```python
num = [5,6,7]
```

```python
filter(lambda item : item%2 == 0 ,num)
```

```python
list(filter(lambda item : item%2 == 0 ,num))
```

```python
list(filter(lambda item : item%2 == 1 ,num))
```
